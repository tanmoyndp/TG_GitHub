### MySQL - Aggregate Functions 
1. AVG() : This function is used to calculate the average value.
2. BIT_AND() : This function is used to calculate the bitwise AND value.
3. BIT_OR() : This function is used to calculate the bitwise OR value.
4. BIT_XOR() : This function is used to calculate the bitwise XOR value.
5. COUNT() : This function is used to count the number of rows returned.
6. GROUP_CONCAT() : This function returns the concatenated string.
7. JSON_ARRAYAGG() : This function returns the given values as a single JSON array.
8. JSON_objectAGG() : This function returns the given values as a single JSON object.
9. MAX() : This function returns the maximum value.
10. MIN() : This function returns the minimum value.
11. STD() : This function calculates and returns the population standard deviation.
12. STDDEV() : This function calculates and returns the population standard deviation.
13. STDDEV_POP() : This function calculates and returns the population standard deviation.
14. STDDEV_SAMP() : This function calculates and returns the sample standard deviation.
15. SUM() : This function is used to calculate the sum of the values.
16. VAR_POP() : This function calculates and returns the population standard variance.
17. VAR_SAMP() : This function calculates and returns the sample variance.
18. VARIANCE : This function calculates and returns the population standard variance.

```python
cursor.execute("""CREATE TABLE MyPlayers(
	ID INT,
	First_Name VARCHAR(255),
	Last_Name VARCHAR(255),
	Socre_In_Exhibiiton_match INT,
	COUNTRY VARCHAR(100),
	PRIMARY KEY (ID)
)""")

cursor.execute("""insert into MyPlayers values (1, 'Shikhar', 'Dhawan', 95, 'India'),
(2, 'Jonathan', 'Trott', 50, 'SouthAfrica'),
(3, 'Kumara', 'Sangakkara', 25, 'Sri Lanka'),
(4, 'Virat', 'Kohli', 50, 'India'),
(5, 'Rohit', 'Sharma', 25, 'India'),
(6, 'Ravindra', 'Jadeja', 15, 'India'),
(7, 'James', 'Anderson', 15, 'England')""")

conn.commit()
```

```python
pd.read_sql_query("""select * from MyPlayers order by COUNTRY
                  """ \
                  , conn)
```

```python
# BIT_AND
pd.read_sql_query("""SELECT ID, First_Name, Last_Name, Country, BIT_AND(Socre_In_Exhibiiton_match) 
                  from MyPlayers group by country
                  """ \
                  , conn)
```

```python
# BIT_OR
pd.read_sql_query("""SELECT ID, First_Name, Last_Name, Country, BIT_OR(Socre_In_Exhibiiton_match) 
                     from MyPlayers group by country
                  """ \
                  , conn)
```

```python
# BIT_xor
pd.read_sql_query("""SELECT ID, First_Name, Last_Name, Country, BIT_xor(Socre_In_Exhibiiton_match) 
                     from MyPlayers group by country
                  """ \
                  , conn)
```

```python
cursor.execute("""CREATE TABLE student (name VARCHAR(15), marks INT, grade CHAR)""")

cursor.execute("""INSERT INTO student VALUES ('Raju', NULL, 'A'),
('Rahman', 60, 'B'),
('Robert', 45, 'C')""")

conn.commit()
```

```python
pd.read_sql_query("""SELECT * from student
                  """ \
                  , conn)
```

```python
# COUNT
pd.read_sql_query("""SELECT COUNT(marks), COUNT(*) from student
                  """ \
                  , conn)
```

```python
# GROUP_CONCAT
pd.read_sql_query("""SELECT COUNTRY, GROUP_CONCAT(Socre_In_Exhibiiton_match),
                     GROUP_CONCAT(First_Name)
                     from MyPlayers
                     group by COUNTRY
                  """ \
                  , conn)
```

```python
# JSON_ARRAYAGG
pd.read_sql_query("""SELECT JSON_ARRAYAGG(Socre_In_Exhibiiton_match) as Scores from MyPlayers
                     group by COUNTRY
                  """ \
                  , conn)
```

```python
# JSON_OBJECTAGG
pd.read_sql_query("""SELECT Country, JSON_OBJECTAGG(First_Name, Socre_In_Exhibiiton_match) as Scores 
                     from MyPlayers GROUP BY country
                  """ \
                  , conn)
```

#### Aggregate function without group by clause

```python
pd.read_sql_query("""SELECT *
                     from productLineSales order by productLine
                  """ \
                  , conn)
```

```python
pd.read_sql_query("""SELECT productLine, min(orderValue) as min_orderValue
                     from productLineSales
                     group by productLine
                  """ \
                  , conn)
```

```python
pd.read_sql_query("""SELECT productLine, min(orderValue) as min_orderValue
                     from productLineSales
                     group by orderYear
                  """ \
                  , conn)
```

```python
pd.read_sql_query("""SELECT
                     productLine, orderYear,
                     min(orderValue) over (partition by productLine) as min_orderValue,
                     max(orderValue) over (partition by productLine) as max_orderValue,
                     JSON_ARRAYAGG(orderValue)  over (partition by productLine order by orderYear
                     RANGE BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) as json_array_orderValue_1,
                     JSON_ARRAYAGG(orderValue)  over (partition by productLine order by orderYear desc
                     RANGE BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) as json_array_orderValue_2,
                     FIRST_VALUE(orderValue) over (partition by productLine order by orderValue) as first_orderValue
                     from productLineSales
                  """ \
                  , conn)
```

```python
# Partition by with different colums
pd.read_sql_query("""SELECT
                     productLine, orderYear,
                     min(orderValue) over (partition by productLine) as min_orderValue,
                     max(orderValue) over (partition by productLine) as max_orderValue,
                     JSON_ARRAYAGG(orderValue)  over (partition by productLine) as json_array_orderValue,
                     FIRST_VALUE(orderValue) over (partition by productLine order by orderValue) as first_orderValue,
                     min(orderValue) over (partition by orderYear) as min_orderValue_Y,
                     max(orderValue) over (partition by orderYear) as max_orderValue_Y,
                     JSON_ARRAYAGG(orderValue)  over (partition by orderYear) as json_array_orderValue_Y
                     from productLineSales
                  """ \
                  , conn)
```
