### Loading sample tables

```python
import pandas as pd
```

```python
party_master = pd.read_excel('data/master_tables.xlsx', engine='openpyxl', sheet_name='party_master')

party_master
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("""CREATE TABLE party_master (
    party_id int,
    party_key varchar(255),
    first_name varchar(255),
    last_name varchar(255),
    address_line_1 varchar(255),
    address_line_2 varchar(255),
    address_line_3 varchar(255),
    zip_code int,
    relationship_start_date date,
    relationship_end_date date,
    customer_type varchar(255),
    status varchar(255)
);""")
```

```python
conn.commit()
```

```python
for index, row in party_master.iterrows():
    str_val = [str(x) for x in row.values]
    values = "', '".join(str_val)
    values = "('" + values + "')"
    qry = """INSERT INTO party_master
             VALUES """ + values
    qry = qry.replace("'NaT'", 'null')
    qry = qry.replace("'nan'", 'null')
    #print(qry)
    cursor.execute(qry)
    conn.commit()
```

```python
transaction_master = pd.read_excel('data/master_tables.xlsx', engine='openpyxl', sheet_name='transaction_master')
transaction_master
```

### Practice Query

```python
pd.read_sql_query("""select pm.*, pal.account_key, pal.account_number, pal.link_type from party_master pm
                  inner join
                  party_account_link pal
                  on pm.party_key = pal.party_key
                  """ \
                  , conn)
```

```python
pd.read_sql_query("""select *
                  from transaction_master
                  """ \
                  , con)
```

```python
pd.read_sql_query("""select year(posting_date) as year, month(posting_date) as month,
                  count(amount_base) as tran_cnt, sum(amount_base) as tran_amt,
                  min(amount_base) as min_amt,
                  NTILE(1000) OVER(ORDER BY amount_base) AS ntile_0,
                  max(amount_base) as max_amt
                  from transaction_master
                  group by
                  year(posting_date), month(posting_date)
                  """ \
                  , con)
```

```python
pd.read_sql_query("""select *
                  from transaction_master order by originator_country, amount_base
                  """ \
                  , con)
```

```python
# Not getting result as intended
pd.read_sql_query("""select originator_country,
                  NTILE(2) OVER(ORDER BY amount_base) AS ntile_3,
                  count(amount_base) as tran_cnt, sum(amount_base) as tran_amt,
                  min(amount_base) as min_amt,
                  NTH_VALUE(amount_base, 1) OVER(partition by originator_country ORDER BY amount_base) as 1_amt,
                  NTH_VALUE(amount_base, 3) OVER(partition by originator_country ORDER BY amount_base) as 2_amt,
                  max(amount_base) as max_amt
                  from transaction_master
                  group by originator_country
                  """ \
                  , con)
```

```python
# Not getting result as intended
pd.read_sql_query("""select originator_country,
                  NTH_VALUE(transaction_key, 1) OVER(partition by originator_country ORDER BY amount_base) as 1_amt,
                  NTH_VALUE(transaction_key, 2) OVER(partition by originator_country ORDER BY amount_base) as 2_amt
                  from transaction_master
                  """ \
                  , con)
```

```python
pd.read_sql_query("""select originator_country,
                  amount_base,
                  DENSE_RANK() OVER(partition by originator_country ORDER BY amount_base) rank_dense,
                  ROW_NUMBER() OVER(partition by originator_country ORDER BY amount_base) rank_row,
                  RANK() OVER(partition by originator_country ORDER BY amount_base) rank_normal,
                  PERCENT_RANK() OVER(partition by originator_country ORDER BY amount_base) rank_percentile
                  from transaction_master
                  order by originator_country, originator_country
                  """ \
                  , conn)
```

```python
pd.read_sql_query("""SELECT DISTINCT originator_country,
                JSON_ARRAYAGG(beneficiary_country) beneficiary_country_s,
                JSON_ARRAYAGG(transaction_key) transaction_key_s
                FROM transaction_master
                group by originator_country
                """ \
                  , conn)
```

```python
pd.read_sql_query("""SELECT DISTINCT originator_country,
                JSON_ARRAYAGG(beneficiary_country) OVER win beneficiary_country_s,
                JSON_ARRAYAGG(transaction_key) OVER win transaction_key_s
                FROM transaction_master
                WINDOW win AS (PARTITION BY originator_country ORDER BY transaction_key)
                """ \
                  , conn)
```
