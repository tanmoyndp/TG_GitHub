### SQLite via Pandas

```python
#import library
import pandas as pd
con = db.connect('sqlite_database.db')
df_employees = pd.read_sql_query('select * from employees', con)
```

```python
df_employees
```

```python
pd.read_sql_query("""select * from employees where
                  empid in (12986, 12987)
                  """ \
                  , con)
```

```python
df_employees.to_sql("employees", con, if_exists="replace")
```

```python
con.close()
```

### Loading sample tables

```python
import pandas as pd
party_master = pd.read_excel('data/master_tables.xlsx', engine='openpyxl', sheet_name='party_master')
```

```python
party_master
```

```python
con = db.connect('sqlite_database.db')
party_master.to_sql("party_master", con, if_exists="replace", index=False)
```

```python
pd.read_sql_query("""select * from party_master
                  """ \
                  , con)
```

```python
account_product_master = pd.read_excel('data/master_tables.xlsx', engine='openpyxl', sheet_name='account_product_master')
account_product_master.to_sql("account_product_master", con, if_exists="replace", index=False)
```

```python
party_account_link = pd.read_excel('data/master_tables.xlsx', engine='openpyxl', sheet_name='party_account_link')
party_account_link.to_sql("party_account_link", con, if_exists="replace", index=False)
```

```python
transaction_master = pd.read_excel('data/master_tables.xlsx', engine='openpyxl', sheet_name='transaction_master')
transaction_master.to_sql("transaction_master", con, if_exists="replace", index=False)
```

### Query on the loaded tables

```python
pd.read_sql_query("""select pm.*, pal.account_key, pal.account_number, pal.link_type from party_master pm
                  inner join
                  party_account_link pal
                  on pm.party_key = pal.party_key
                  """ \
                  , con)
```

```python
pd.read_sql_query("""select * from transaction_master
                  """ \
                  , con)
```

```python
pd.read_sql_query("""select strftime('%Y', posting_date) as year, strftime('%m', posting_date) as month,
                  count(amount_base) as tran_cnt, sum(amount_base) as tran_amt,
                  min(amount_base) as min_amt, max(amount_base) as max_amt
                  from transaction_master
                  group by
                  strftime('%Y', posting_date), strftime('%m', posting_date)
                  """ \
                  , con)
```
